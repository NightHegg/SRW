## Создание сетки и триангуляция области

```python
from itertools import product
from scipy.spatial import Delaunay
import numpy as np
par = 4
bounds=np.array([[0.01, 0], [0.02, 0], [0.02, 0.01], [0.01, 0.01], [0.01, 0]])
x_list, y_list=np.linspace(0.01, 0.02, par), np.linspace(0, 0.01, par)
points = np.array(list(product(x_list,y_list)))
tri = Delaunay(points)
dirichlet_conditions = [[0, 1], [1, 0], [3, 0]]
neumann_conditions = [[2, 0, -2e+7]]

dirichletPoints = [[[idx for idx, val in enumerate(points) 
                     if min([bounds[cond[0], 0], bounds[cond[0] + 1, 0]]) <= val[0] <= max([bounds[cond[0], 0], bounds[cond[0] + 1, 0]])
                     and min([bounds[cond[0], 1], bounds[cond[0] + 1, 1]]) <= val[1] <= max([bounds[cond[0], 1], bounds[cond[0] + 1, 1]])], cond[1]] for cond in dirichlet_conditions]
neumannPoints = [[[idx for idx, val in enumerate(points) 
                     if min([bounds[cond[0], 0], bounds[cond[0] + 1, 0]]) <= val[0] <= max([bounds[cond[0], 0], bounds[cond[0] + 1, 0]])
                     and min([bounds[cond[0], 1], bounds[cond[0] + 1, 1]]) <= val[1] <= max([bounds[cond[0], 1], bounds[cond[0] + 1, 1]])], cond[1], cond[2]] for cond in neumann_conditions]

```

## Построение сетки

## Создание файла с данными

```python
#Коэффициенты для задачи
dimTask=2
amntSubdomains = [2, 1]
coefOverlap=0.3
E=70e+9
nyu=0.34
coef_u=1000
coef_sigma=1e-3
```

```python
def CalculateStiffnessMatrix(element, points):
    B=np.zeros((3, 6))
    x_points=np.array([points[element[0], 0], points[element[1], 0], points[element[2], 0]])
    y_points=np.array([points[element[0], 1], points[element[1], 1], points[element[2], 1]])
    a=np.array([x_points[1] * y_points[2] - x_points[2] * y_points[1], 
               x_points[2] * y_points[0] - x_points[0] * y_points[2],
               x_points[0] * y_points[1] - x_points[1] * y_points[0]])
    b=np.array([y_points[1]-y_points[2], y_points[2]-y_points[0], y_points[0]-y_points[1]])
    c=np.array([x_points[2]-x_points[1], x_points[0]-x_points[2], x_points[1]-x_points[0]])
    A=0.5*np.linalg.det(np.array([[1, 1, 1], [x_points[0], x_points[1], x_points[2]], 
                                  [y_points[0], y_points[1], y_points[2]]]).transpose())
    for i in range(len(element)):
        B[:, i*dimTask:i*dimTask + 2] += np.array([[b[i], 0], [0, c[i]], [c[i]/2, b[i]/2]])/2/A
    return B, A

def boundCondition_Neumann(F, element, neumannPoints_Local, dimTask, value, points, dim):
    L = list(set(element) & set(neumannPoints_Local))
    len = np.linalg.norm(np.array(points[L[0]]) - np.array(points[L[1]]))
    for node in L:
        F[node * dimTask + dim] += value * len / 2
    return F

def boundCondition_Dirichlet(K, F, dimTask, node, value, dim):
    for i in range(len(F) // dimTask):
        F[i * dimTask + dim] -= K[i * dimTask + dim][node * dimTask + dim] * value
                
    K[node * dimTask + dim][:] = 0
    K[:][node * dimTask + dim] = 0
    
    K[node * dimTask + dim][node * dimTask + dim] = 1
    F[node * dimTask + dim] = value
    return K, F
```

```python
import numpy as np
import scipy
from itertools import combinations
K_global = np.zeros((len(points) * 2, len(points) * 2))
F=np.zeros((len(points) * 2, 1))
D=np.array([[1, nyu/(1-nyu), 0], [nyu/(1-nyu), 1, 0], [0, 0, (1-2 * nyu) / 2 / (1-nyu)]]) * E * (1-nyu) / (1-2 * nyu) / (1+nyu)
for element in tri.simplices:
    B, A = CalculateStiffnessMatrix(element, points)
    K=np.dot(np.dot(B.transpose(), D), B) * A
    for i in range(3):
        for j in range(3):
            K_global[element[i]*dimTask:element[i]*dimTask+2, element[j]*dimTask:element[j]*dimTask+2]+=K[i*dimTask:i*dimTask+2,j*dimTask:j*dimTask+2]
            
for condition in dirichletPoints:
    for node in condition[0]:
        K_global, F = boundCondition_Dirichlet(K_global, F, dimTask, node, 0, condition[1])
        
for condition in neumannPoints:
    for element in [element for element in tri.simplices for x in list(combinations(condition[0], 2)) if x[0] in element and x[1] in element]:
        F = boundCondition_Neumann(F, element.tolist(), condition[0], dimTask, condition[1], points, 0)
        F = boundCondition_Neumann(F, element.tolist(), condition[0], dimTask, condition[2], points, 1) 
u=np.linalg.solve(K_global, F).reshape((points.shape[0],2)) * coef_u
print(u)
```

```python
plt.triplot(points[:,0], points[:,1], tri.simplices.copy())
plt.triplot(u[:, 0] + points[:, 0], u[:,1] + points[:, 1], tri.simplices.copy())
plt.plot(points[:,0] + u[:, 0], points[:,1] + u[:, 1], 'o')
plt.show()
```

```python
Eps=[]
for element in tri.simplices:
    B, A = CalculateStiffnessMatrix(element, points)
    Eps.append(np.dot(B, np.ravel(np.array([u[element[0]], u[element[1]], u[element[2]]])).transpose()))
Eps=np.array(Eps)
Sigma=np.dot(D, Eps.transpose())
print(Sigma.transpose()*coef_sigma)
```

```python
import numpy as np
from itertools import groupby
relation_PointsElements_GS = {}
boundPoints = []
subdElements_GS = []
sortedSubdPoints_GS = []
sortedSubdPointsCoords_GS = []
subdBoundary = []
subdBoundary_Schwarz = []

subdBounds = [bounds[0][0]] + [bounds[0][0] + i*(bounds[1][0] - bounds[0][0])/amntSubdomains[0] for i in range(1, amntSubdomains[0])] + [bounds[1][0]]
overlapBounds = [subdBounds[0]] + sum([[subdBounds[x] - (subdBounds[x] - subdBounds[x - 1]) * coefOverlap, subdBounds[x] + (subdBounds[x] - subdBounds[x - 1]) * coefOverlap]   
               for x in range(1, len(subdBounds)-1)], []) + [subdBounds[-1]]
subdElements_GS.append([element.tolist() for element in tri.simplices 
                         if overlapBounds[0] < sum([points[element[0], 0], points[element[1], 0], points[element[2], 0]])/3 < overlapBounds[2]])
for i in range(1, len(overlapBounds) - 3, 2):
    subdElements_GS.append([element.tolist() for element in tri.simplices 
                         if overlapBounds[i] < sum([points[element[0], 0], points[element[1], 0], points[element[2], 0]])/3 < overlapBounds[i+3]])
subdElements_GS.append([element.tolist() for element in tri.simplices 
                         if overlapBounds[-3] < sum([points[element[0], 0], points[element[1], 0], points[element[2], 0]])/3 < overlapBounds[-1]])

for subd in subdElements_GS:
    sortedSubdPoints_GS.append([el for el, _ in groupby(sorted(sum([i for i in subd], [])))])

for subd in sortedSubdPoints_GS:
    sortedSubdPointsCoords_GS.append(np.array([points[i].tolist() for i in subd]))

for idx, val in enumerate(points):
    relation_PointsElements_GS[idx] = [element.tolist() for element in tri.simplices if idx in element.tolist()]
    
boundaryPoints = [idx for idx, val in enumerate(points) if val[0] in [bounds[0, 0], bounds[1, 0]] or val[1] in [bounds[0, 1], bounds[2, 1]]]

for idv, subd in enumerate(subdElements_GS):
    tempList = []
    relation_PointsElements_subd = {}
    for point in sortedSubdPoints_GS[idv]:
        relation_PointsElements_subd[point] = [element for element in subd if point in element]
    for idx, val in relation_PointsElements_subd.items():
        if val != relation_PointsElements_GS[idx]:
            tempList.append(idx)
    #for point in boundaryPoints:
    #    if point in sortedSubdPoints_GS[idv]:
    #        tempList.append(point)
    subdBoundary.append(tempList)
    
subdBoundary_Schwarz.append(list(set(subdBoundary[0]) & set(sortedSubdPoints_GS[1])))
for i in range(1, len(subdElements_GS) - 1):
    subdBoundary_Schwarz.append(list((set(subdBoundary[i]) & set(sortedSubdPoints_GS[i + 1])) | (set(subdBoundary[i]) & set(sortedSubdPoints_GS[i - 1]))))
subdBoundary_Schwarz.append(list(set(subdBoundary[-1]) & set(sortedSubdPoints_GS[-2])))
```

```python
it = 0
u_global = np.zeros((points.shape[0], 2))
D=np.array([[1, nyu/(1-nyu), 0], [nyu/(1-nyu), 1, 0], [0, 0, (1-2 * nyu) / 2 / (1-nyu)]]) * E * (1-nyu) / (1-2 * nyu) / (1+nyu)
while True:
    u_Previous = np.copy(u_global)
    for idv, subd in enumerate(subdElements_GS):
        
        ratioPoints_LocalGlobal = dict(zip(range(len(sortedSubdPoints_GS[idv])), sortedSubdPoints_GS[idv]))
        ratioPoints_GlobalLocal = {v: k for k, v in ratioPoints_LocalGlobal.items()}
        
        K_global = np.zeros((len(ratioPoints_GlobalLocal) * 2, len(ratioPoints_GlobalLocal) * 2))
        F = np.zeros((len(ratioPoints_GlobalLocal) * 2, 1))
        subdElements_Local = np.array([ratioPoints_GlobalLocal[x] for x in np.array(subd).ravel()]).reshape(len(subd), 3)

        for element in subdElements_Local:
            B, A = CalculateStiffnessMatrix(element, sortedSubdPointsCoords_GS[idv])
            K = np.dot(np.dot(B.transpose(), D), B) * A
            for i in range(3):
                for j in range(3):
                    K_global[element[i]*dimTask:element[i]*dimTask+2, element[j]*dimTask:element[j]*dimTask+2] += K[i*dimTask:i*dimTask+2,j*dimTask:j*dimTask+2]

        for condition in dirichletPoints:
            for node in list(set(condition[0]) & set(sortedSubdPoints_GS[idv])):
                K_global, F = boundCondition_Dirichlet(K_global, F, dimTask, ratioPoints_GlobalLocal[node], 0, condition[1])
        
        for condition in neumannPoints:
            for element in [element for element in subd for x in list(combinations(condition[0], 2)) if x[0] in element and x[1] in element]:                
                F = boundCondition_Neumann(F, [ratioPoints_GlobalLocal[x] for x in element], [ratioPoints_GlobalLocal[x] for x in list(set(sortedSubdPoints_GS[idv]) & set(condition[0]))], dimTask, condition[1], sortedSubdPointsCoords_GS[idv], 0)
                F = boundCondition_Neumann(F, [ratioPoints_GlobalLocal[x] for x in element], [ratioPoints_GlobalLocal[x] for x in list(set(sortedSubdPoints_GS[idv]) & set(condition[0]))], dimTask, condition[2], sortedSubdPointsCoords_GS[idv], 1)
        
        #for node in subdBoundary[idv]:
           # K_global, F = boundCondition_Dirichlet(K_global, F, dimTask, ratioPoints_GlobalLocal[node], u_global[node, 0], dim = 0)
            #K_global, F = boundCondition_Dirichlet(K_global, F, dimTask, ratioPoints_GlobalLocal[node], u_global[node, 1], dim = 1)
        for node in list(set(subdBoundary[idv]) - set(neumannPoints[0][0])):
            K_global, F = boundCondition_Dirichlet(K_global, F, dimTask, ratioPoints_GlobalLocal[node], u_global[node, 1], dim = 1)
            K_global, F = boundCondition_Dirichlet(K_global, F, dimTask, ratioPoints_GlobalLocal[node], u_global[node, 0], dim = 0)
           
        u_local = np.linalg.solve(K_global, F).reshape((sortedSubdPointsCoords_GS[idv].shape[0], 2))     
        
        print(u_local)
        for x in list(ratioPoints_LocalGlobal.keys()):
            u_global[ratioPoints_LocalGlobal[x], :] = u_local[x, :]

        plt.triplot(points[:,0], points[:,1], tri.simplices.copy())
        plt.triplot(u_global[:, 0] * coef_u + points[:, 0], u_global[:,1] * coef_u + points[:, 1], tri.simplices.copy())
        plt.plot(points[:,0] + u_global[:, 0] * coef_u, points[:,1] + u_global[:, 1] * coef_u, 'o')
        plt.show()
    it += 1
    if it == 5:
        break
u_global = u_global * coef_u
print(u_global)
```
