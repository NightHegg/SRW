## Задание 5
Что такое спектральный радиус матрицы R? Приведите условие сходимости
итерационного процесса в терминах спектрального радиуса. Приведите критерии
сходимости методов Якоби, Зейделя и последовательной верхней релаксации для
различных классов матриц.

Спектральным радиусом матрицы $R$ называется число
\begin{equation}
\rho(R) = max_i \left | \lambda_i \right |,
\end{equation}
где $\lambda_i$ - собственные значения матрицы $R$. 

Условие сходимости итерационного процесса в терминах спектрального радиуса:

Итерационный процесс $x_{m+1} = Rx_m + c$ сходится к решению системы $Ax=b$, каков бы ни был начальный вектор $\Leftrightarrow$ $\rho(R)<1. 

Критерии сходимости:
- метод Якоби: $\rho(R_j)<1$, где $R_j = L + U$;
- метод Зейделя: $\rho(R_{GS})<1$, где $R_{GS} = (I-L)^{-1} U$; 
- последовательная верхняя релаксация: $\rho(R_{SOR})<1$, где $R_{SOR} = (I-\omega L)^{-1} ((1-\omega)I + \omega U)$, ($\omega >1$);

## Задание 7
Что такое подпространство Крылова для матрицы A и вектора b? Что такое
алгоритм Ланцоша и какую информацию он позволяет получить на k -м шаге? Из
каких соображений можно строить алгоритмы с помощью крыловского
подпространства? Какая характеристика матрицы А влияет на сходимость
итерационного процесса в методе сопряженных градиентов? Как можно ускорить
процесс сходимости?

Подпространство Крылова $K_k(A, b)$ - линейная оболочка векторов $b, Ab, A^2 b$, ... $A^{k-1} b$.

Алгоритм Ланцоша - алгоритм для "частичного" приведения матрицы к симметричной трехдиагональной форме. На k-м шаге мы получим всю достаточную информацию для получения решения системы $Ax=b$.
